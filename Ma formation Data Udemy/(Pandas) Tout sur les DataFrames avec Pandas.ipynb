## Tout sur les DataFrames avec Pandas

## Index partagé par toutes les colonnes

Les objets df peuvent intéragir facilement avec de nombreuses colonnes.

Chacune de ces colonnes est représentée par un objet Series.
Les objets Series gardent l'alignement entre valeurs et leurs intitulés d'index.

Pour les df qui sont une collection d'objets Series, ils gardent l'alignement entre colonnes et lignes.

Pandas partage un index de ligne pour plusieurs colonnes
Par défaut, cet index est représenté par des entiers (int)

On a donc un index partagé par toutes les colonnes. Résultat si on essaye de lire un fichier csv contenant un nombre d'éléments différents par colonne, cela retourne une erreur.

```python
import pandas as pd
```

## Utiliser des index personnalisés

```python
# DataFrame.set_index(Series, inplace=True, drop=False)
#Par défaut, Pandas va créer un nouveau df et l'indexer avec les valeurs de la colonne spécifiée
# puis supprime la colonne utilisée pour le nouvel index.
# - Series : correspondra au nouvel index
# - inplace : si on spécifie la valeur true cela ne créera pas de nouveau df et ça va remplacer l'index du df d'origine par le nouveau correspondant à la colonne spécifiée en paramètre "Séries"
# - drop : la colonne spécifiée en paramètre "Series" sera conservée dans le df au lieu d'être supprimée.
```

* Utiliser la méthode set_index pour indéxer notre DataFrame fandango avec la colonne "FILM". A préciser que nous souhaitons conserver la colonne "FILM". De plus nous souhaitons garder le DataFrame d'origine et en créer un nouveau que nous assignerons à la variable fandango_films.
* Afficher l'index du DataFrame fandango_films à l'aide de l'attribut index.

```python
fandango_films = fandango.set_index("FILM", drop=False)
```

```python
print(fandango_films.index)
```

## Sélectionner des valeurs depuis un index personnalisé

```python
# sélection avec crochets ou méthode loc[]
fandango_films["Mr. Holmes (2015)":"Two Days, One Night (2014)"]
```

```python
fandango_films.loc["Mr. Holmes (2015)":"Two Days, One Night (2014)"]
```

```python
#film spécifique
fandango_films.loc["Kumiko, The Treasure Hunter (2015)"]
```

```python
# Sélection depuis une liste de films
movies = ["Kumiko, The Treasure Hunter (2015)", "Danny Collins (2015)", "Ant-Man (2015)"]
fandango_films.loc[movies]
```

Lorsque nous sélectionnons des lignes multiples, Pandas retourne un df. Et lorsque nous sélectionnons une seule ligne, Pandas retourne un objet Series à la place.

Dans les deux cas, Pandas conserve l'index d'origine, avec des entiers, même si nous avons spécifié un nouvel index. Ce qui signifie que l'on peut toujours sélectionner les lignes par les numéros (par les entiers, par l'ancien index)

## Méthode Apply() sur les colonnes d'un DataFrame

```python
# apply()
# permet d'appliquer une fonction sur les objets Series d'un df
```

```python
# dataframe.apply(lambda x: x**2)
# cela retournerait un nouveau df en ayant élevé à la puissance 2 chaque élément des colonnes numériques
```

```python
import numpy as np
```

```python
#retourner les types de données en tant que Series
```

```python
types = fandango_films.dtypes
```

```python
types
```

```python
# filtrer les nombres à virgules float64 et retourner le nom des colonnes
float_columns = types[types.values == "float64"].index
float_columns
```

```python
# utilisation des crochets pour filtrer les colonnes de la condition float_columns
float_df = fandango_films[float_columns]
float_df
```

```python
deviations = float_df.apply(lambda x: np.std(x))
```

```python
#On applique la fonction std avec la méthode apply() sur le df float_df, qui va calculer l'écart type de chaque colonne
# dans l'annotation le x est un objet Series et correspond à chaque colonne
# ce qui signifie : applique pour chaque x la fonction std de calcule de l'écart type.
# cela retourne un objet Series contenant les écarts types pour chaque colonne
```

```python
print(deviations)
```

* Obtenir le dataframe float_df que j'ai réalisé en exemple (qui retourne les colonnes décimales float64)
* Utiliser la méthode apply() sur ce dataframe float_df afin de diviser par 2 chaque valeur et assigner le résultat à la variable exo_df.
* Afficher les premières lignes du résultat.

```python
types = fandango_films.dtypes
float_columns = types[types.values == "float64"].index
float_df = fandango_films[float_columns]
```

```python
exo_df = float_df.apply(lambda x: x/2)
print(exo_df)
```

## Méthode Apply() sur les lignes d'un DataFrame

```python
# pour appliquer une fonction sur toutes les lignes d'un df, il peut utiliser le paramètre axis et le fixer à 1
```

```python
# apply(function, axis=1)
```

```python
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
rt_mt_user
```

```python
rt_mt_user.apply(lambda x: np.std(x), axis=1)
```

   * Utiliser la méthode apply() pour calculer la moyenne (méthode mean() de la librarie Numpy) des valeurs de chaque film pour "RT_user_norm" et "metacritic_user_nom" puis assigner le résultat à la varible rt_mt_means.
   * Afficher les 5 premières valeurs de rt_mt_ means.

```python
import numpy as np
```

```python
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
```

```python
rt_mt_means = rt_mt_user.apply(lambda x: np.mean(x), axis=1)
print(rt_mt_means[0:5])
```
