```python
# importer librairies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importer dataset
dataset = pd.read_csv("regression polynomiale/Position_Salaries.csv")
```

```python
dataset
```

```python
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

```python
# Construction du modèle
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x, y)
```

```python
# Visualiser les résultats
plt.scatter(x, y, color= "red")
plt.plot(x, regressor.predict(x), color= "blue")
plt.title("Salaire vs Expérience")
plt.xlabel("Expérience")
plt.ylabel("Salaire")
plt.show()
```

## Création d'une courbe régression Polynomiale

```python
# importer librairies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importer dataset
dataset_poly = pd.read_csv("regression polynomiale/Position_Salaries.csv")
```

```python
x = dataset_poly.iloc[:, 1:2].values
y = dataset_poly.iloc[:, -1].values
```

```python
# Construction du modèle
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

```python
regressor = LinearRegression()
regressor.fit(x_poly, y)
```

```python
# Visualiser les résultats
plt.scatter(x, y, color= "red")
plt.plot(x, regressor.predict(x_poly), color= "blue")
plt.title("Salaire vs Expérience")
plt.xlabel("Expérience")
plt.ylabel("Salaire")
plt.show()
```
